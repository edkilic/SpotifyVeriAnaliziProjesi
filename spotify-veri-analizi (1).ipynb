```python
def ms_to_minsec(ms):
    sec = ms / 1000
    return f"{int(sec // 60)}:{int(sec % 60)}"


sureler = df[['duration_ms', 'year']].groupby('year').mean().reset_index().iloc[1:-1]

sureler['duration_s'] = sureler['duration_ms'] / 1000

sureler['min:sec'] = sureler['duration_ms'].apply(ms_to_minsec)

iplot(px.line(sureler, 
        x='year', 
        y='duration_s',
        title='Yıllara göre ortalama şarkı süresi',
        text='min:sec'
       ).update_xaxes(type='category').update_traces(textposition='top right'))

##Bu kod, şarkı sürelerinin yıllar içinde nasıl değiştiğini gösteren bir çizgi grafiği oluşturur.
```

```python
histogram_etiketleri = ['popularity',
                    'danceability', 
                    'energy', 
                    'speechiness', 
                    'loudness', 
                    'acousticness', 
                    'liveness', 
                    'instrumentalness',
                    'valence',
                    'tempo'
                   ]

colors = px.colors.qualitative.Vivid
for i in range(len(histogram_etiketleri)):
    fig = px.histogram(df, 
                       histogram_etiketleri[i], 
                       title=f'{histogram_etiketleri[i]} en iyi hitlerdeki dağılımı', 
                       color_discrete_sequence=[colors[i]]
                      )
    iplot(fig)
##Bu kod, şarkıların farklı özelliklerinin (popülarite, dans edilebilirlik, enerji vb.) dağılımını gösteren histogramlar oluşturur.
```
